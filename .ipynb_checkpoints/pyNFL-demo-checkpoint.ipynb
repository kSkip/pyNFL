#Example 2. Query
Below is a demostration of the kind of direct information that can be extracted about individual players from the enclosed database named,"nfl.db". In the first cell we connect to the sqlite3 database and construct a query that requests all records of the player's name, team, and receiving yards of all wide receivers and tight ends that played **away** against the New England Patriots. Then we construct a Pandas DataFrame object out of the result

````python
conn = sqlite3.connect('nfl.db')
c = conn.cursor()

c.execute('SELECT team, name, yds_receiving FROM nfl_stats where (position=\'WR\' OR position=\'TE\') AND (opponent = \'@NE\'OR opponent = \'NE\') AND year > 2013')

df = pd.DataFrame(c.fetchall(),columns=['team','name','yards'])

conn.close()
````

Now we use grouby() and aggregate() methods to organize a new DataFrame consisting of each player with their total accumulated yards from all the records. sort() and head() will give us only the top ten by yards. So now we can see, "Who does the best at Gillette Stadium?"

````python
tds_total = df.groupby('name').aggregate(sum).sort('yards',ascending=False).head(10)

print(tds_total)

tds_total.plot(kind='bar',title='top ten WR & TE receving yards against NE 2014-2015',legend=False)
````

#Example 3. Extrapolation
"What is the best QB-WR combo in league by touchdowns?" This question was posed to me by a friend, who questioned the utility of this library (this information is easily googled). I immediately thought that this is solved with a simple SQL query. On a second thought, the information pulled from nfl.com does not specify who threw the touchdown pass and who received the pass. Fear not! as there is a solution (at least partial), and is a great demostration of the power of extrapolating information from such a reduced data set. The point is not that you can answer just the popular questions, but that you can answer your own questions or others'.

One might think, ok you rank the number of touchdown passes received by all the WRs. The SQL looks like this:
```sql
SELECT name, td_receiving FROM nfl_stats;
```
Then you cast the result to a Pandas DataFrame and groupby and aggregate the rows to produce the totals of all the players:
```python
df.groupby('name').aggregate(sum)
```
Nice, except that this assumes the same QB plays every single game, the whole game. This is commonly the case. However, an injured quarterback made skew the results, or players may change teams. So, you need to find the QB that did throw the TDs in each week.

To throw a wrench in the mix, the data from nfl.com regarding the player's team, is the player's **current** team, and not the team they played for in that week. The fix is to find combinations of QBs and WRs that played *against* the same team in the same week. This implies they were on the same team and the specific team is not necessarily too relavent to the question at hand.

The last complication is that we cannot determine the combo if there were two QBs in the same game that both made touchdown passes. Since this is not very common, I will just throw out those out because it's not going to significantly affect the statistics.

Here we go! First, we need to pull the data from the database.

````python
conn = sqlite3.connect('nfl.db')
c = conn.cursor()

c.execute('SELECT week, name, td_receiving, td_passing, position, opponent FROM nfl_stats where (position=\'WR\' OR position=\'QB\') AND year = 2015')

qb_wr_df = pd.DataFrame(c.fetchall(),columns=['week','name','receiving TDs','passing TDs','position','opponent'])

conn.close()
````

We groupby *week* and *opponent*, so we can collect the names of players who where on the same team

````python

qb_wr_groups = qb_wr_df.groupby(['week','opponent'],as_index=False)
````

Now, we loop through the groups and pull out the QB-WR combos and store them in a DataFrame 

````python
QB_WR_Combos = dict()

for team_week in qb_wr_groups:
    
    the_qbs = team_week[1]["passing TDs"] > 0
    if the_qbs.sum() == 1:
        
        #get the QB's name
        the_qb = team_week[1][the_qbs]['name'].to_string().split('   ')[1]

        #which WRs scored
        the_wrs = team_week[1]["receiving TDs"] > 0
        
        #if at least one WR scored
        if the_wrs.sum() > 0:
            
            #get the WRs into a new DataFrame and convert to strings
            team_wr_data = pd.DataFrame(team_week[1][the_wrs][['name','receiving TDs']])
            team_wr_strings = team_wr_data.to_string(header=False,index=False).split('\n')
            
            for wr in team_wr_strings:
                
                name_tds = re.match(r'([A-Za-z\. -]*)([0-9]*)',wr.strip())
                wr_name = name_tds.group(1)
                wr_tds = int(name_tds.group(2))

                combo = the_qb.strip() + ' & ' + wr_name.strip()
                if combo in QB_WR_Combos.keys():
                    QB_WR_Combos[combo] += wr_tds
                else:
                    QB_WR_Combos[combo] = wr_tds
````

Now, that we have a dictionary full of the combos with their respective number of shared touchdowns, we can plot the data in the bar chart and see whos best! We will plot only the top 10 combos, so *sort* and *head* will be useful.

As you can see, for the 2015 season we have a three way tie between Eli & Odell, Blake & Allen, and Russell & Doug (Go Hawks!)

````python
#convert to DataFrame
combo_df = pd.DataFrame(list(QB_WR_Combos.items()),columns=['Combo','TDs'])

#obtain the top 10
top_10_combos = combo_df.sort('TDs',ascending=False).head(10)

#plot
top_10_combos.plot(kind='bar',x="Combo",y="TDs",legend=False)
````
